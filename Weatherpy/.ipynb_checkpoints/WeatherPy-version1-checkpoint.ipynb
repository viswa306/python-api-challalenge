```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint
from datetime import date
# Import API key
#from api_keys import weather_api_key
from config1 import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
pprint(cities)
len(cities)
```

```python
# Save config information.
url =  "http://api.openweathermap.org/data/2.5/weather?"

# Build partial query URL
units="metric"
query_url = f"{url}appid={weather_api_key}&q="
print(query_url)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
 #city_re=requests.get(query_url + city+ "&units=" + units).json()
```

```python
humidity=[]
counter = 0
print(f'Beginning Data Retrieval')  
print(f'-----------------------------')
for city in cities:
    counter = counter + 1
   # response=requests.get(query_url + city).json()
    response=requests.get(query_url + city+ "&units=" + units).json()
    #pprint(response)
    try: 
   # pprint(response)
        #humidity= response["main"]["humidity"]
         #lat.append(response["coord"]["lat"])
        print(f'Processing Record {counter} of set 1 |{city}' ) 
    except:
            print(f'City not found. Skipping...:{city}')
            pass
print(f'-----------------------------')   
print(f'Ending of Data Retrieval')  
print(f'-----------------------------')
```

```python
lat=[]
lng=[]
max_temp=[]
humidity=[]
cloudiness=[]
windspeed=[]
Country=[]
Date=[]
cityname=[]
cityid=[]
counter= 0

for city in cities:
    response=requests.get(query_url + city+"&units=" + units).json()
    counter = counter + 1
    try: 
  
           #humidity= response["main"]["humidity"]
           lat.append(response["coord"]["lat"])
           lng.append(response["coord"]["lon"])
           humidity.append(response["main"]["humidity"])
           max_temp.append(response["main"]["temp_max"]) 
           cloudiness.append(response["clouds"]["all"]) 
           windspeed.append(response["wind"]["speed"])
           Country.append(response["sys"]["country"])
           Date.append(response["dt"])
           cityname.append(response["name"])
           cityid.append(counter)
            
           #cityid.append(response["id"]) 
        
    except:
            #print(f'latitude not found. Skipping...:{city}')
            pass
#print(f'latitude :{lat}') 
#len(lat)
#print(f'humidity:{humidity}')
#len(humidity)
#print(f'logitude:{lng}')
#print(f'Maximum temperature:{max_temp}')
#print(f'Clodiness:{cloudiness}')
#print(f'wind speed:{ windspeed}')
# print(f'Country:{Country}')
# print(f'Date:{Date}')
#print(f'cityid:{cityid}')
#print(f'cityname:{cityname}')
#print(f'cityid:{cityid}')
```

```python
Raw_dict={"Cityid":cityid,"City":cityname,"Cloudiness":cloudiness,"Country":Country,"Date":Date,"Humidity":humidity,
                      "latitude":lat,"Longitude":lng,"Maxtemp":max_temp,"Windspeed":windspeed}
raw_data=pd.DataFrame(Raw_dict)
raw_data
```

```python
#finding the outliers for the Humidity 
quartiles = raw_data['Humidity'].quantile([.25,.5,.75])

lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq

print(f"The lower quartile of Humidity is: {lowerq}")
print(f"The upper quartile of Humidity is: {upperq}")
print(f"The interquartile range of Humidity is: {iqr}")
print(f"The the median of Humidity is: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
print(f"Values below {lower_bound} could be outliers for Humidity.")
print(f"Values above {upper_bound} could be outliers for Humidity.")
```

```python
## Make a new DataFrame equal to the city data to drop all humidity outliers by index.

city_data=raw_data.loc[(raw_data["Humidity"] < lower_bound) | (raw_data["Humidity"] > upper_bound)]
city_data
```

```python
# need to check:potential outlier >lower bound and lessthan upperbound which meansabove 14 and below 130
city_data=raw_data.loc[(raw_data["Humidity"] > lower_bound) &(raw_data["Humidity"]< upper_bound)]
city_data
```

```python
# Make a new DataFrame equal to the city data to drop all humidity outliers by index.
city_data_df=raw_data.loc[(raw_data["Humidity"] > lower_bound) &(raw_data["Humidity"]< upper_bound)]
city_data_df

# Passing "inplace=False" will make a copy of the city_data DataFrame, which we call "clean_city_data".
clean_city_data= city_data_df.reset_index(drop=True,inplace=False)
clean_city_data
```

```python
# Extract relevant fields from the data frame
# Reorganizing the columns using double brackets

City_Data = clean_city_data[["City","latitude","Longitude","Maxtemp","Humidity","Cloudiness","Windspeed","Country","Date"]]
City_Data 
# Export the City_Data into a csv
```

## Linear Regression

```python
# Add the linear regression equation and line to plot for latitude vs windspeed
x_values =  City_Data['latitude']
y_values = City_Data['Windspeed']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.title('latitude vs windspeed')
plt.xlabel('Latitude')
plt.ylabel('wind speed')
plt.show()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Northern Hemisphere:Add the linear regression equation and line to plot for Max Temp vs. Latitude 

x_values = North_hemisphere['latitude']
y_values =North_hemisphere  ['Maxtemp']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(40,30),fontsize=15,color="red")
plt.title('Northern Hemisphere - Max Temp vs. Latitude')
plt.xlabel('latitude')
plt.ylabel('Maxtemperature')

print(f"The r-squared is: {rvalue**2}")
plt.show()
```
