### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Store latitude and longitude in locations
locations = City_data_df[["latitude","Longitude"]]


Humidity=City_data_df["Humidity"]
```

```python
#plot heatmap
fig= gmaps.figure()
#create heat layer
heat_layer = gmaps.heatmap_layer(locations,weights=Humidity,
                                 dissipating=False,
                                 max_intensity=10,
                                 point_radius=1)
#Add layer
fig.add_layer(heat_layer)

#Display figure
fig
```

```python
# BONUS
# Plost as a Hybrid map
fig = gmaps.figure(map_type="HYBRID")

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations,weights=Humidity,
                                 dissipating=False,
                                 max_intensity=10,
                                 point_radius=1)

fig.add_layer(heat_layer)
fig
```

```python
# BONUS
# Plot as a Terrain map
fig = gmaps.figure(map_type="TERRAIN")

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations,weights=Humidity,
                                 dissipating=False,
                                 max_intensity=10,
                                 point_radius=1)

fig.add_layer(heat_layer)
fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df=pd.DataFrame(ideal_weather)
hotel_df
```

```python
# Add columns for "Hotel Name"
hotel_df["Hotel Name"] = ""
df4=hotel_df.head(10)
df4
```

```python
from pprint import pprint
```

```python
# params dictionary to update each iteration
params = {
    
    "radius": 50000,
    "keyword": "hotels",
    "key": g_key
}
# Use the lat/lng we recovered to identify hotels
for index, row in df4.iterrows():
    # get lat, lng from df
    lat = row["latitude"]
    lng = row["Longitude"]
 # change location each iteration while leaving original params in place
    params["location"] = f"{lat},{lng}"   
     # Use the search term: "Hotel" and our lat/lng
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # make request and print url
    response = requests.get(base_url, params=params)
#print(name_address)
     # convert to json
    hotel_response= response.json()
    try:
             hotel_df.loc[index, "Hotel Name"] =  hotel_response["results"][0]["name"]
       # hotel_df.loc[index, "Airport Address"] = name_address["results"][0]["vicinity"]
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
    
#pprint(h_d)
df5=hotel_df.head(10)       
df5
```
